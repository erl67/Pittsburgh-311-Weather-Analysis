This program will read a csv file of 311 data to a dataframe.
The dataframe will be cleaned as such: Remove all data other than the year 2017

The results will be stored to a new csv file.

```python
import pandas as pd
import numpy as np
import datetime as datetime
import time
```

```python
df = pd.read_csv('76fda9d0-69be-4dd5-8108-0de7907fc5a4.csv')
```

```python
df.head(3), df.tail(3)
```

```python
df.describe()
```

```python
df = df[df['CREATED_ON'].str.contains('2017-')]
df = df.sort_values(ascending=False, axis=0, by=['REQUEST_ID'])
df.reset_index(drop=True, inplace=True)
df.describe()
```

```python
#should be ~14MB cleaned to 2017
with open('311cleaned.csv', 'w') as fh:
    df.to_csv(fh, index=False)
```
